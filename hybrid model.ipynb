# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix,f1_score
```

```python
# Define different values for max_depth
max_depth_values = [1, 5, 10, 20, 30]
```

```python
# Initialize lists to store performance metrics
accuracies = []
precisions = []
recalls = []
specificities = []
```

```python
# Train and evaluate the decision tree classifier for each max_depth value
for depth in max_depth_values:
    # Initialize the decision tree classifier with the current max_depth
    dt_classifier = DecisionTreeClassifier(max_depth=depth, random_state=42)
    
    # Train the classifier
    dt_classifier.fit(X_train, y_train)
    
    # Predict the target variable on the test set
    y_pred = dt_classifier.predict(X_test)
    
    # Calculate performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    
    # Calculate specificity
    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    specificity = tn / (tn + fp)
    
    # Append the metrics to the lists
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    specificities.append(specificity) 
```

```python
# Print the results
for i, depth in enumerate(max_depth_values):              
    print(f"Max Depth: {depth}")
    print(f"Accuracy: {accuracies[i]}")
    print(f"Precision: {precisions[i]}")
    print(f"Recall: {recalls[i]}")
    print(f"Specificity: {specificities[i]}")
    print()
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Define the function to calculate specificity
def specificity_score(y_true, y_pred):
    tn, fp, _, _ = confusion_matrix(y_true, y_pred).ravel()
    return tn / (tn + fp)
```

```python
# Define different values for max_iter
max_iter_values = [10, 20, 30, 40, 50]
```

```python
# Perform experiments with different values of n_neighbors
for n_neighbors in [2, 3, 4]:
    # Create and train the K-Neighbors Classifier
    knn_model = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn_model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = knn_model.predict(X_test)
    
    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    specificity = specificity_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    
    # Print the results
    print(f"Results for n_neighbors = {n_neighbors}:")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"Specificity: {specificity:.2f}")
    print(f"F1-score: {f1:.2f}")
    print()
```

# RF+DT+XGB

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import VotingClassifier
from sklearn.cluster import FeatureAgglomeration
```

```python
# Define the classifiers
rf = RandomForestClassifier()
xgb = GradientBoostingClassifier()
dt = DecisionTreeClassifier()
```

```python
# Create the pipeline with Canopy-based feature selection
pipeline = Pipeline([
    ('feature_selection', FeatureAgglomeration()),
    ('voting', VotingClassifier(estimators=[('rf', rf), ('xgb', xgb), ('dt', dt)]))
])
```

```python
# Define the parameter grid for grid search
param_grid = {
    'voting__voting': ['soft', 'hard'],
    'feature_selection__n_clusters': [10, 20, 30]
}
```

```python
# Perform grid search
grid_search = GridSearchCV(pipeline, param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
# Get the best model
best_model = grid_search.best_estimator_
```

```python
# Fit the best model
best_model.fit(X_train, y_train)
```

```python
# Evaluate the performance
y_pred = best_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```

# RF+XGB+ada

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectFromModel
from sklearn.cluster import FeatureAgglomeration
from sklearn.metrics import accuracy_score
```

```python
# Define the base estimators
rf_clf = RandomForestClassifier()
xgb_clf = GradientBoostingClassifier()
dt_clf = DecisionTreeClassifier()
ada_clf = AdaBoostClassifier(estimator=dt_clf)  # Use 'estimator' instead of 'base_estimator'
```

```python
# Create a pipeline with feature selection and ensemble classifier
pipeline = Pipeline([
    ('feature_selection', FeatureAgglomeration()),
    ('classification', VotingClassifier(estimators=[
        ('rf', rf_clf), 
        ('xgb', xgb_clf), 
        ('ada', ada_clf)], voting='soft'))
])
```

```python
# Define the parameter grid for grid search
param_grid = {
    'feature_selection__n_clusters': [5, 10, 15],
    'classification__ada__n_estimators': [50, 100, 150],
    'classification__rf__n_estimators': [50, 100, 150],
    'classification__xgb__n_estimators': [50, 100, 150]
}
```

```python
# Perform grid search with cross-validation
grid_search = GridSearchCV(pipeline, param_grid, cv=5)
grid_search.fit(X_train, y_train) 
```

```python
# Get the best parameters and estimator from the grid search
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_
```

```python
# Make predictions on the test set using the best estimator
y_pred = best_estimator.predict(X_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Best Parameters:", best_params)
print("Accuracy:", accuracy)
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
